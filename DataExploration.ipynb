```python
def lang(text):
    try:
        DetectorFactory.seed = 0
        return detect(text)
    except:
        return None
```

```python
stock_bot['language'] = stock_bot['text'].progress_apply(lambda x: lang(x))
```

```python
CustomerKey = ""
CustomerSecretKey = ""
AccessToken = ""
AccessTokenSecret = ""
```

```python
# Creating the authentication object
auth = tweepy.OAuthHandler(CustomerKey, CustomerSecretKey)

# Setting your access token and secret
auth.set_access_token(AccessToken, AccessTokenSecret)

# Creating the API object while passing in auth information
api = tweepy.API(auth,wait_on_rate_limit=True)
# #
# row_list = [["created_at", "Tweet_id","full_text", "entities","geo","source","user_id_str","user_name","user_location","user_description","user_followers_count","user_friends_count",
#              "user_created_at","user_verified","user_statuses_count","favorited","favorite_count","retweet_count","reply_count","profile_image_url_https","keyword_contained"]]

row_list = []

# The search term you want to find
# Language code (follows ISO 639-1 standards)
language = "en"

queries = ["BTC"]

# queries = ["‡∏ä‡∏≤‡∏ï‡∏¥‡∏ö‡πâ‡∏≤‡∏ô‡πÄ‡∏°‡∏∑‡∏≠‡∏á","‡πÅ‡∏ú‡πà‡∏ô‡∏î‡∏¥‡∏ô","#‡∏≠‡∏ô‡∏∏‡∏ä‡∏ô‡∏£‡∏±‡∏Å‡∏ä‡∏≤‡∏ï‡∏¥‡∏®‡∏≤‡∏™‡∏ô‡πå‡∏Å‡∏©‡∏±‡∏ï‡∏£‡∏¥‡∏¢‡πå"]
# Calling the user_timeline function with our parameters
# results = api.search(q=query, lang=language,tweet_mode='extended')

for query in queries:
    max_tweets = 2000
    searched_tweets = [status for status in tweepy.Cursor(api.search, q=query,lang=language,tweet_mode='extended').items(max_tweets)]
    
   # foreach through all tweets pulled
    for tweet in tqdm(searched_tweets):
       # printing the full_text stored inside the tweet object
       if (not tweet.retweeted) and ('RT @' not in tweet.full_text):
           if ("reply_count" not in tweet._json):
               tweetlist = [tweet.created_at, tweet.id_str, tweet.full_text, tweet.entities,tweet.geo,tweet.source,
                            tweet.user.id_str, tweet.user.name, tweet.user.location,tweet.user.description,
                            tweet.user.followers_count, tweet.user.friends_count,
                            tweet.user.created_at, tweet.user.verified,
                            tweet.user.statuses_count, tweet.favorited,
                            tweet.favorite_count, tweet.retweet_count, 0,
                            tweet.user.profile_image_url_https, query]
           else:
               tweetlist = [tweet.created_at,tweet.id_str, tweet.full_text, tweet.entities,tweet.geo,tweet.source,
                            tweet.user.id_str, tweet.user.name, tweet.user.location,tweet.user.description,
                            tweet.user.followers_count, tweet.user.friends_count,
                            tweet.user.created_at, tweet.user.verified,
                            tweet.user.statuses_count, tweet.favorited,
                            tweet.favorite_count, tweet.retweet_count, tweet.reply_count,
                            tweet.user.profile_image_url_https, query]
           if tweetlist not in row_list:
               row_list.append(tweetlist)

with open('TwitterDB_BTC.csv', 'a', newline='') as file:
   writer = csv.writer(file)
   writer.writerows(row_list)
```

```python
btc = btc[btc['user_followers_count'].apply(lambda x: x.isnumeric())]
btc['user_followers_count'] = btc['user_followers_count'].astype('float64')
btc['user_followers_level'] = pd.qcut(btc['user_followers_count'], 5, labels=["low", "slightly medium", "medium","slightly high", "high"])

```

```python
high_btc = btc[btc['user_followers_level']=='high']
high_btc = high_btc[['Tweet_id','created_at','full_text','user_id_str','user_name',
          'user_followers_count','user_followers_level']]
```

```python
high_btc.to_csv('HighFollowBTC.csv',index=False)
```
